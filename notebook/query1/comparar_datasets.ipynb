# COMPARAR DATASETS


```python
import pandas as pd

csv_a = "../../output/client_output.csv"
csv_b = "transacciones_usd.csv"

def comparar_csv_sin_orden(archivo1, archivo2):
    try:
        # 1. Leer los archivos CSV
        df1 = pd.read_csv(archivo1)
        df2 = pd.read_csv(archivo2)
        
        # 2. Verificar que tengan las mismas columnas (aunque estén en diferente orden)
        if set(df1.columns) != set(df2.columns):
            print("❌ Los archivos NO son iguales: Tienen columnas diferentes.")
            print(f"Columnas CSV 1: {list(df1.columns)}")
            print(f"Columnas CSV 2: {list(df2.columns)}")
            return False

        # 3. Reordenar las columnas de df2 para que coincidan con el orden de df1
        df2 = df2[df1.columns]

        # 4. Ordenar las filas de ambos dataframes por todos sus valores
        #    Esto garantiza que el orden de las filas ya no afecte la comparación
        df1_ordenado = df1.sort_values(by=list(df1.columns)).reset_index(drop=True)
        df2_ordenado = df2.sort_values(by=list(df1.columns)).reset_index(drop=True)

        # 5. Comparar si los DataFrames resultantes son idénticos
        if df1_ordenado.equals(df2_ordenado):
            print("✅ ¡Los archivos CSV son exactamente IGUALES! (ignorando el orden)")
            return "Son Iguales"
        else:
            print("❌ Los archivos NO son iguales: El contenido de las filas difiere.")
            return "Son Diferentes"

    except FileNotFoundError as e:
        print(f"❌ Error: No se pudo encontrar uno de los archivos: {e}")
    except Exception as exc:
        print(f"❌ Ocurrió un error inesperado: {exc}")

comparar_csv_sin_orden(csv_a, csv_b)
```
