Cuenta de origen, cuenta de destino y monto para transacciones USD menores a 50.

```python
# importamos dataset
import numpy as np
import pandas as pd

CANTIDAD_FILAS = 10000
RUTA_DE_DATASETS = "../../datasets"
NOMBRE_ARCHIVO = "LI-Small_Trans.csv"
# NOMBRE_ARCHIVO = "transacciones_sample.csv"
transacciones = pd.read_csv(f"{RUTA_DE_DATASETS}/{NOMBRE_ARCHIVO}").sample(n=CANTIDAD_FILAS)
transacciones.head()

def guardar_csv(df, nombre_archivo):
    df.to_csv(nombre_archivo, index=False)
guardar_csv(transacciones, f"{RUTA_DE_DATASETS}/transacciones_sample.csv")
```

```python
#Filter USD transactions.
transacciones_usd = transacciones[transacciones['Payment Currency'] == "US Dollar"]
guardar_csv(transacciones_usd, "transacciones_usd.csv")
```

```python
# filtro por fecha 01/09 al 06/09
transacciones_septiembre_1_al_5 = transacciones_usd[(transacciones_usd["Timestamp"] >= '2022/09/01') & (transacciones_usd["Timestamp"] <= '2022/09/06')]
guardar_csv(transacciones_septiembre_1_al_5, "transacciones_septiembre_1_al_5.csv")
```

```python
# filtro por monto
MONTO = 50  

transacciones_menores_50 = transacciones_usd[transacciones_usd['Amount Paid'] < MONTO]
guardar_csv(transacciones_menores_50, "transacciones_menores_50.csv")
```
