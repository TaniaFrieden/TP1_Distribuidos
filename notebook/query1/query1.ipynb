# QUERY 1
Cuenta de origen, cuenta de destino y monto para transacciones USD menores a 50.

```python
# importamos dataset
import numpy as np
import pandas as pd

RUTA_DE_DATASETS = "../../datasets"
# NOMBRE_ARCHIVO = "trans_sample.csv"
NOMBRE_ARCHIVO = "transacciones_sample_q4.csv"

SOLUCION = "q1_solucion.csv"
transacciones = pd.read_csv(f"{RUTA_DE_DATASETS}/{NOMBRE_ARCHIVO}")

def guardar_csv(df, nombre_archivo):
    df.to_csv(nombre_archivo, index=False)
```

```python
transacciones_usd = transacciones[transacciones['Payment Currency'] == "US Dollar"]
MONTO = 50  
transacciones_menores_50 = transacciones_usd[transacciones_usd['Amount Paid'] < MONTO]
resultado = transacciones_menores_50[['From Bank', 'Account', 'To Bank','Account.1', 'Amount Paid']]
guardar_csv(resultado, SOLUCION)
```
