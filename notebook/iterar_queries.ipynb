```python
import importlib
import utils.obtener_queries as obtener_queries_module
from utils.comparar_datasets import comparar_csv_sin_orden
from utils.titular import imprimir_titulo


ACTUAL_TEMPLATE = "../output/output_{q}.csv"
EXPECTED_TEMPLATE = "./query{q}/q{q}_solucion.csv"

CANTIDAD_EJECUCIONES = 3

obtener_queries_module = importlib.reload(obtener_queries_module)
QUERIES = obtener_queries_module.obtener_queries_desde_compose()
fallas_por_query = {q: 0 for q in QUERIES}

for indice in range(1, CANTIDAD_EJECUCIONES + 1):
    imprimir_titulo(f"Iteración {indice}/{CANTIDAD_EJECUCIONES}")
    
    !make -C .. client !logs

    for query in QUERIES:
        actual_csv = ACTUAL_TEMPLATE.format(q=query)        
        expected_csv = EXPECTED_TEMPLATE.format(q=query)
        print(f"\nComparando CSVs para Query {query}:")
        son_iguales, mensaje = comparar_csv_sin_orden(actual_csv, expected_csv)
        print(mensaje)

        if son_iguales:
            print("Resultado notebook: iguales")
            continue

        fallas_por_query[query] += 1
        print("Resultado notebook: diferentes")

imprimir_titulo("Resumen")

print("\nResumen de fallas por query:")
for query, fallas in fallas_por_query.items():
    print(f"Query {query}: {fallas} fallas en {CANTIDAD_EJECUCIONES} ejecuciones")
```
